**TASK 1 (TITANIC SURVIVAL PREDICTION)**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic_df = pd.read_csv('/content/titanic.csv')
titanic_df
```

```python
titanic_df.head(10)
```

```python
titanic_df.tail(10)
```

```python
titanic_df.describe()
```

```python
titanic_df.shape
```

```python
titanic_df.info()
```

```python
sns.heatmap(titanic_df.isnull())
```

```python
ports = pd.get_dummies(titanic_df.Embarked,prefix='Embarked',dtype=int)
ports
```

```python
ports.head()
```

```python
titanic_df = titanic_df.join(ports)
titanic_df
```

```python
titanic_df.drop(['Embarked'],axis=1,inplace=True)
```

```python
titanic_df.head()
```

```python
titanic_df.Sex = titanic_df.Sex.map({'male':0,'female':1})
```

```python
titanic_df['Sex']
```

```python
Y = titanic_df.Survived.copy()
X = titanic_df.drop(['Survived'],axis=1)
```

```python
X.drop(['Name','Ticket','Cabin','PassengerId'],axis=1,inplace=True)
```

```python
X.isnull().sum()
```

```python
X.Age.fillna(X.Age.mean(),inplace=True)
```

```python
X.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=7)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,Y_train)
```

```python
model.score(X_train,Y_train)
```

```python
Y_pred_test = model.predict(X_test)
Y_pred_test
```

```python
Y_test
```

```python
from sklearn.metrics import accuracy_score
print('model accuracy score: {0:0.4f}'.format(accuracy_score(Y_test,Y_pred_test)))
```

TASK - 2 (IRIS FLOWER  CLASSIFICATION)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
tk2_df =pd.read_csv('/content/Iris.csv')
tk2_df
```

```python
tk2_df.head(5)
```

```python
tk2_df.tail(5)
```

```python
tk2_df['Species'].value_counts()
```

```python
sns.pairplot(tk2_df,hue='Species')
```

```python
X = tk2_df[["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]]
y = tk2_df["Species"]
```

```python
print(X)
print(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
X_train
```

```python
X_test
```

```python
ML = KNeighborsClassifier(n_neighbors=10)
ML.fit(X_train, y_train)
```

```python
X_train_prediction = ML.predict(X_train)
print(X_train_prediction)
```

```python
print(y_train)
```

```python
X_test_prediction = ML.predict(X_test)
print(X_test_prediction)
```

```python
print(y_test)
```

```python
train_accuracy = accuracy_score(X_train_prediction, y_train)
test_accuracy = accuracy_score(X_test_prediction, y_test)
print("Accuracy scores of traning and test data are", train_accuracy, 'and' , test_accuracy, 'respectively')
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_train, X_train_prediction))
```

```python
print(classification_report(y_test, X_test_prediction))
```
